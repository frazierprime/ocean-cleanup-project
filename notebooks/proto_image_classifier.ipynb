```python
import matplotlib.pyplot as plt
import seaborn as sns

import torch
from torchvision.models.efficientnet import EfficientNet, MBConvConfig


import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D , MaxPool2D , Flatten , Dropout 
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam

from sklearn.metrics import classification_report,confusion_matrix

import tensorflow as tf

import cv2
import os

import numpy as np
```

```python
from ocean_cleanup_project import utils
```

```python
DEVICE = "cpu"
classes = utils.OCEAN_CLEANUP_CLASSES
classes
```

```python
path_to_model = '/workspace/.cache/kili/automl/clb6jq0rp0f560lyef4n15v57/CLASSIFICATION_JOB/torchvision/efficientnet_b0'
sample_image_path = '/workspace/data/test/100001.jpg'
```

```python
# load our canned model
model = utils.get_canned_efficientnet_model()
```

```python
# Initialize the image transforms.
transform = utils.preprocess()

# prep test image
cv2_image = cv2.imread(sample_image_path)
rgb_image = cv2.cvtColor(cv2_image, cv2.COLOR_BGR2RGB)
# apply transformations to image.
input_tensor = transform(cv2_image)

input_batch = input_tensor.unsqueeze(0)

# Send input tensor and model to the computation device.
input_batch = input_batch.to(DEVICE)
model.to(DEVICE)
```

```python
import time
with torch.no_grad():
    start_time = time.time()
    output = model(input_batch)
    end_time = time.time()

# Get the softmax probabilities.
probabilities = torch.nn.functional.softmax(output[0], dim=0)
# Check the top 4 classes that are predicted.
top5_prob, top5_catid = torch.topk(probabilities, 4)
print(classes)
for i in range(top5_prob.size(0)):
    cv2.putText(cv2_image, f"{top5_prob[i].item()*100:.3f}%", (15, (i+1)*30), 
                cv2.FONT_HERSHEY_SIMPLEX,
                1, (0, 0, 255), 2, cv2.LINE_AA)
    # cv2.putText(cv2_image, f"{classes[top5_catid[i]]}", (160, (i+1)*30), 
    #             cv2.FONT_HERSHEY_SIMPLEX,
    #             1, (0, 0, 255), 2, cv2.LINE_AA)
    print(top5_prob[i].item())
cv2.imshow('Result', cv2_image)
cv2.waitKey(0)
# Define the outfile file name.
save_name = f"outputs/{args['input'].split('/')[-1].split('.')[0]}_{DEVICE}.jpg"
cv2.imwrite(save_name, cv2_image)
print(f"Forward pass time: {(end_time-start_time):.3f} seconds")
```
